```python
air_quality= air_quality.replace(to_replace=-200, value=np.nan)
```

```python
air_quality.isnull().sum()
```

```python
# 1. Calculate the mean for numeric columns only
mean_values = air_quality.mean(numeric_only=True)

# 2. Fill the missing values and assign it back to the variable
air_quality = air_quality.fillna(mean_values)

# 3. Check the result
air_quality.tail()
```

```python
# 1. Convert Date with dayfirst=True to handle the DD/MM/YYYY format
air_quality['Date'] = pd.to_datetime(air_quality['Date'], dayfirst=True)

# 2. Replace the dots in Time with colons and convert to timedelta
air_quality['Time'] = pd.to_timedelta(air_quality['Time'].str.replace('.', ':'))

# 3. Combine them into a single DateTime column for easier plotting
air_quality['DateTime'] = air_quality['Date'] + air_quality['Time']

# 4. Check the result
air_quality[['Date', 'Time', 'DateTime']].head()
```
